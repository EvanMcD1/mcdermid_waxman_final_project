```python
#game_date takes the team and year, and returns home_game_dates, away_game_dates, game_result.
def game_date(year, team):
    #load the specific year
    df = pd.read_csv(f"{path}/Season_Scores/{year}_scores.csv")

    #subsets the df to games where the team is playing
    df_away = df[(df["AwayTeam"] == team)]
    df_home = df[(df["HomeTeam"] == team)]

    print(df_away, df_home)
    
    home_game_dates = df_home['Date'].to_list()
    away_game_dates = df_away['Date'].to_list()
    home_game_win = df_home['HomeWin'].to_list()
    away_game_win = df_away['HomeWin'].to_list()

    #Convert to boolean, while preserving the nan so we know to skip bye week
    home_game_win = [bool(x) if pd.notna(x) else x for x in home_game_win if x is not None]
    away_game_win = [bool(x) if pd.notna(x) else x for x in away_game_win if x is not None]

    home = list(zip(home_game_dates, home_game_win))
    away = list(zip(away_game_dates, away_game_win))
    
    
    return home, away
```
